```python
def smile_mood_block(user_text: str) -> str:
    """
    SMILE ka basic mood analysis:
    - Mood word
    - Mood score 1â€“10
    - Emoji
    - Emotional aura (color + short meaning)
    """
    prompt = f"""
You are SMILE â€” a gentle emotional wellness companion.

User message:
"{user_text}"

TASK:
1. Detect the user's mood in 1â€“2 words.
2. Give a mood score from 1 to 10 (1 = very low, 10 = very good).
3. Give ONE emoji that matches the mood.
4. Give an emotional aura in this format:
   Aura Color: <color-name or emoji>
   Aura Meaning: <1 short line>

FORMAT YOUR ANSWER LIKE THIS EXACTLY:

Mood: <mood word>
Mood Score: <number>/10
Emoji: <one emoji>
Aura Color: <color or emoji>
Aura Meaning: <short explanation>
"""
    response = model.generate_content(prompt)
    return response.text
```

```python
def smile_voice_reply(user_text: str, voice: str = "friend") -> str:
    """
    voice options:
    - 'friend'
    - 'parent'
    - 'teacher'
    - 'mentor'
    - 'god'
    - 'future-self'
    """
    prompt = f"""
You are SMILE â€” an emotional wellness companion.

User's message:
"{user_text}"

User wants comfort in the voice of: {voice}

VOICE RULES:
- friend  = casual, yaari wali tone, thoda fun + caring
- parent  = warm, protective, "beta" type loving support
- teacher = firm but caring, growth + discipline focus
- mentor  = guidance + clarity, future-focused
- god     = spiritual, peaceful, reassuring
- future-self = talk like their older, stronger, wiser self

Write 3â€“5 lines max.
Use Hindi/English/Hinglish mix naturally based on the sample,
but keep tone SOFT and HELPFUL.
Include 1â€“2 gentle emojis.
"""
    response = model.generate_content(prompt)
    return response.text
```

```python
def smile_healing_task(user_text: str) -> str:
    """
    User ki current state ke hisaab se 
    ek chhota sa healing task deta hai (2â€“5 min).
    """
    prompt = f"""
You are SMILE, an emotional wellness agent.

User said:
"{user_text}"

TASK:
Give a SHORT healing micro-task that the user can do in 2â€“5 minutes.

RULES:
- Use soft Hinglish (Hindi + English mix).
- Make it very simple and realistic.
- Focus on: breathing, grounding, gratitude, self-compassion, or mini movement.
- Use bullet points or steps (max 4 steps).
- End with 1 short encouragement line.

FORMAT:

Healing Task:
- Step 1: ...
- Step 2: ...
- Step 3: ...

Closing line: <1 line with emoji>
"""
    response = model.generate_content(prompt)
    return response.text
```

```python
def smile_rescue_mode(user_text: str) -> str:
    """
    Jab user bohot anxious / overwhelmed feel kare,
    yeh quick 3â€“4 step calm-down protocol deta hai.
    """
    prompt = f"""
You are SMILE â€” an emotional first-aid coach.

User message (high stress):
"{user_text}"

TASK:
Create a 3â€“4 step EMERGENCY CALM-DOWN routine.

RULES:
- Start with: "ðŸš¨ Rescue Mode: Let's calm down together."
- Step 1: Breathing (simple count)
- Step 2: Grounding using senses (5-4-3-2-1 or similar)
- Step 3: Replace one negative thought with a kinder thought
- Step 4: 1 very small self-care action (water, stretching, etc.)
- Use soft Hinglish, short steps.

FORMAT:

ðŸš¨ Rescue Mode: Let's calm down together ðŸŒ¿

Step 1: ...
Step 2: ...
Step 3: ...
Step 4: ...

Closing line: <short reassurance with emoji>
"""
    response = model.generate_content(prompt)
    return response.text
```

```python
mood_log = []  # global list: har entry ek dict hogi

def log_mood_entry(day_label: str, mood_score: int, emoji: str, note: str):
    """
    mood_log me daily entry store karega.
    """
    entry = {
        "day": day_label,
        "score": mood_score,
        "emoji": emoji,
        "note": note
    }
    mood_log.append(entry)
```

```python
def SMILE_full_flow(user_text: str, voice: str = "friend"):
    """
    SMILE Agent (Final Version)
    - Short replies for greeting
    - Medium replies for neutral messages
    - Full emotional support for sad/anxious messages
    - Uses 4 custom voice personalities (Friend / Parent / God / Future-Self)
    - Automatic English/Hindi/Hinglish support
    """

    # Lowercase message for keyword matching
    msg = user_text.lower().strip()

    # SHORT MESSAGE LIST (1â€“2 line reply)
    short_msgs = ["hi", "hello", "hey", "hlo", "hm", "hmm", "ok", "okay"]

    # --- SHORT REPLY ---
    if msg in short_msgs:
        prompt = f"""
You are SMILE, an emotional wellness companion.
User: "{user_text}"

Respond in 1â€“2 lines only.
Tone: soft, friendly, simple.
Do NOT give mood, tasks, or long paragraphs.
Language: auto-detect (Hindi/English/Hinglish)
Emoji: 1 soft emoji at end.
"""
        return model.generate_content(prompt).text

    # MEDIUM KEYWORDS (normal feelings)
    medium_keys = ["tired", "confused", "fine", "theek", "normal", "okay", "thoda"]

    if any(k in msg for k in medium_keys):
        prompt = f"""
You are SMILE.
User: "{user_text}"

Reply in 3â€“4 lines.
Tone: caring, soft, practical.
Include:
- 1 mood guess
- 1 small advice
- NO inner child
- NO rescue mode
- NO long healing tasks
Language: auto-detect.
Emoji: 1 soft emoji.
"""
        return model.generate_content(prompt).text

    # EMOTIONAL KEYWORDS (full support)
    emotional_keys = [
        "sad", "anxious", "stress", "cry", "broken", "hurt",
        "overthink", "scared", "fear", "lonely", "panic",
        "nhi ho raha", "khatam", "give up", "lost"
    ]

    # FINAL 4 VOICE PERSONALITIES
    personality = {
        "friend": """
TONE: Chill, funny, supportive friend.
STYLE:
- Uses light jokes
- Keeps situation light
- Gives simple + correct advice
- Casual Hinglish/English
- Uses "yaar", "bro", "relax" vibes
""",
        "parent": """
TONE: Loving, warm parent.
STYLE:
- Calls user "beta", "bacha"
- Gives emotional safety
- Encourages growth gently
- Soft Hindi/English mix
""",
        "god": """
TONE: Spiritual + wise divine guide.
STYLE:
- Calls user "Arjun"
- Focus on karma, solutions, clarity
- Gives courage + direction
- Calm & powerful tone
""",
        "future-self": """
TONE: Wise older version of user.
STYLE:
- Motivates from the future
- Reassures: "We win later"
- Mature, hopeful tone
- English/Hinglish comfort
"""
    }

    # --- FULL EMOTIONAL SUPPORT ---
    if any(k in msg for k in emotional_keys):
        prompt = f"""
You are SMILE, an emotional wellness agent.

User message:
"{user_text}"

VOICE MODE: {voice.upper()}
VOICE PERSONALITY:
{personality[voice]}

Write a full emotional support reply containing:

1. Mood (1â€“2 words)
2. Mood Score (1â€“10)
3. Emoji

4. Comfort message (in {voice} style)

5. Healing Task:
   - ONLY 2â€“3 short steps
   - Simple, practical

6. Rescue Mode:
   - ONLY IF mood score <= 4
   - short 2â€“3 steps

7. Inner Child:
   - ONLY IF mood score <= 4
   - max 3 soft lines

Tone Rules:
- Use Hinglish/Hindi/English automatically based on user text
- No long paragraphs
- Clean, readable format
"""

        return model.generate_content(prompt).text

    # DEFAULT MEDIUM REPLY
    prompt = f"""
User: "{user_text}"

Give a 3â€“4 line caring reply.
Small advice + 1 emoji.
Language: auto.
"""
    return model.generate_content(prompt).text
```

```python
def SMILE_chat():
    print("ðŸŒ¿ Welcome to SMILE â€” Your Emotional Wellness Companion ðŸŒ¿\n")
    print("Type 'exit' to stop chatting.\n")
    
    while True:
        user_msg = input("You: ")

        if user_msg.lower() in ["exit", "quit", "bye"]:
            print("\nSMILE: Take care Anshika ðŸ’› Main yahin hoon jab bhi chaho. ðŸŒ¿")
            break

        # â­ THIS IS THE IMPORTANT PART â­
        voice = choose_voice()     # <<â€” THIS CALLS THE MENU

        # Full response
        reply = SMILE_full_flow(user_msg, voice=voice)
        
        print("\nSMILE:\n")
        print(reply)
        print("\n" + "-"*60 + "\n")
```

```python
def choose_voice():
    print("\nWhose voice do you want comfort from?")
    print("1. Friend")
    print("2. Parent")
    print("3. God")
    print("4. Future-Self\n")
    
    choice = input("Select (1â€“6): ").strip()
    
    voices = {
        "1": "friend",
        "2": "parent",
        "3": "god",
        "4": "future-self"
    }
    
    # Default friend if invalid choice
    return voices.get(choice, "friend")

```

```python
mood_log = []   # store all chats
def log_mood_auto(mood_score, emoji, text):
    entry = {
        "mood_score": mood_score,
        "emoji": emoji,
        "text": text
    }
    mood_log.append(entry)
```

```python
import re

def extract_mood_details(smile_output):
    # Example line: "Mood Score: 4/10"
    score_match = re.search(r"Mood Score:\s*(\d+)", smile_output)
    emoji_match = re.search(r"Emoji:\s*([^\n]+)", smile_output)

    mood_score = int(score_match.group(1)) if score_match else None
    emoji = emoji_match.group(1).strip() if emoji_match else None
    
    return mood_score, emoji
```

```python
import matplotlib.pyplot as plt

def plot_mood_graph():
    if not mood_log:
        print("No mood entries to plot yet!")
        return
    
    scores = [entry["mood_score"] for entry in mood_log]
    emojis = [entry["emoji"] for entry in mood_log]
    x = list(range(1, len(scores) + 1))

    plt.figure(figsize=(7,4))
    plt.plot(x, scores, marker="o", linestyle="-")
    plt.xticks(x, emojis)  # emojis shown on xâ€“axis
    plt.ylim(0, 10)
    plt.title("SMILE Diary â€“ Mood Trend")
    plt.xlabel("Chat Entries")
    plt.ylabel("Mood Score (1â€“10)")
    plt.grid(True)
    plt.show()
```
